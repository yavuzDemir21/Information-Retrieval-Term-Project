```python
Habitat_Trie = {}
```

```python
def construct_habitat(root, word_list, i, obt_id):
    if i == len(word_list):
        root['is_habitat'] = True
        root['OBT'] = obt_id
        return root
    elif word_list[i] == '':
        i = i+1
        construct_habitat(root, word_list, i, obt_id)
    else:
        word = word_list[i]
        if(word not in root):
            root[word] = {}
    
        i = i+1
        construct_habitat(root[word], word_list, i, obt_id)
    
    return root
    
```

```python
import os
```

```python
BB_norms = os.listdir("./BioNLP-OST-2019_BB-norm_train/")
```

```python
for txt in BB_norms:
    if txt[len(txt)-2:] == 'a1':
        f = open("BioNLP-OST-2019_BB-norm_train/"+txt, "r")
        file_text = f.read()
        tag_list = file_text.split('\n')
        
        a2_name = txt.replace('a1', 'a2')
        
        f_a2 = open("BioNLP-OST-2019_BB-norm_train/"+a2_name, "r")
        a2_text = f_a2.read()
        obt_list = a2_text.split('\n')
        obt_list.remove('')

        obt_dict = {}
        for obt in obt_list:
            obt_tokens = obt.split(' ')
            tag = obt_tokens[1][obt_tokens[1].index(':')+1:]
            obt_id = obt_tokens[2][obt_tokens[2].rindex(':')+1:]
            if tag not in obt_dict:
                obt_dict[tag] = obt_id
            else:
                obt_dict[tag] = obt_dict[tag]+','+obt_id
        
        for tag in tag_list:
            if 'Habitat' in tag:
                cur_habitat = tag[tag.rindex('\t')+1:]
                words_in_cur = cur_habitat.split(' ')
                
                tag_id = tag[0:tag.index('\t')]
                obt_id = obt_dict[tag_id]
                
                construct_habitat(Habitat_Trie, words_in_cur, 0, obt_id)
    
```
